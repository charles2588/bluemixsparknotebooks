```python
#Insert / Replace your credentials here for your Azure SQL / Synapse Database
Azure_SQL_Creds = {
    'username': 'admin@chuckdatabase',
    'password': """XXXXX""",
    'database': 'chuckdatabase',
    'host': 'chuckgomestest.database.windows.net',
    'port': '1433',
}
```

```python
# Form the connectionstring for your Azure SQL database or your sql server database
# Note you may need to modify this based on the authentication, the below should work for SQL authentication and may need more parameters for Azure AD authentication
# https://docs.microsoft.com/en-us/sql/connect/jdbc/connecting-using-azure-active-directory-authentication?view=sql-server-ver15
connectionstring = "jdbc:sqlserver://" + Azure_SQL_Creds["host"]+":"+Azure_SQL_Creds["port"]+";database="+Azure_SQL_Creds["database"]+";encrypt=true;trustServerCertificate=false;hostNameInCertificate=*.database.windows.net;loginTimeout=30;"            
```

```python
import jaydebeapi
conn = jaydebeapi.connect(
"com.microsoft.sqlserver.jdbc.SQLServerDriver",
connectionstring,
[Azure_SQL_Creds["username"], Azure_SQL_Creds["password"]],"mssql-jdbc-8.2.0.jre8.jar")
```

```python
mapd_cursor = conn.cursor()
```

```python
query = "select * from sys.tables"
mapd_cursor.execute(query)
results = mapd_cursor.fetchall()
```

```python
import pandas as pd
df = pd.DataFrame(results)
```
